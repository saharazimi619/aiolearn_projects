```python
image=cv.imread('pic1.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
face=model.detectMultiScale(image)
```

```python
image=cv.imread('pic1.jpg')
model=cv.CascadeClassifier('model.xml')

face=model.detectMultiScale(image)

x=face[0][0]
y=face[0][1]
a=face[0][2]
b=face[0][3]

image=cv.rectangle(image,(x,y),(x+a,y+b),(0,255,0),4)
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
plt.imshow(image)
```
